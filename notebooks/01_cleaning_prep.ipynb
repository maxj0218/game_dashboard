Now that i have the data formatted and the columns i want, i can start cleaning

```python
selected_data = raw_data.copy()
selected_data.head()
```

```python
# putting release date in datetime format

selected_data['Release date'] = pd.to_datetime(selected_data['Release date'], errors='coerce')
print(selected_data['Release date'])
```

```python
#cleaning out all games with 0-0 estimated owners
selected_data = selected_data[selected_data['Estimated owners'] != '0 - 0']
print(selected_data['Estimated owners'])
```

```python
#creating midpoint column for owners, and sorting df by that midpoint
owner_split = selected_data['Estimated owners'].str.split(' - ', expand=True)

selected_data["owners_low"] = pd.to_numeric(owner_split[0])
selected_data["owners_high"] = pd.to_numeric(owner_split[1])

selected_data["owners_mid"] = (selected_data["owners_low"] + selected_data["owners_high"]) / 2
```

```python
sorted_owner_data = selected_data.sort_values(by = 'owners_mid')
sorted_owner_data.tail(10)
```
