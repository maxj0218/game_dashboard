# Time Series Analysis

Analyze trends in game success over time.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('../data/processed/cleaned_games.csv')
df['Release date'] = pd.to_datetime(df['Release date'])
df['year'] = df['Release date'].dt.year

# Average owners by year
yearly_owners = df.groupby('year')['owners_mid'].mean()
yearly_owners.plot(figsize=(10,6))
plt.title('Average Owners by Release Year')
plt.show()
```
